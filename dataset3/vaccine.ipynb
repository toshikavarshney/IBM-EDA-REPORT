```python
#Renaming the Columns

df = df.rename(columns={df.columns[0]:'Date',
          df.columns[7]:'Male',
          df.columns[8]:'Female',
          df.columns[9]:'Transgender',
          df.columns[10]:'Covaxin'})
df.head()
```

```python
#Plotting the Bar Graph 
df1 = df.head()
import matplotlib.pyplot as plt
plt.bar(df1.Date,df1.Covaxin,color=['r'])
plt.xlabel('Date')
plt.ylabel('Covaxin Doses')
plt.title('Bar Graph')
plt.show()
```

```python
#Scatter Plot 
df2 = df.head()
plt.scatter(df2['Date'],df2['Covaxin'])
plt.show()
```

```python
#Line Graph

df1 = df.groupby('Date').aggregate({'Covaxin':'mean'})
df1.plot(figsize=(15,3))
```

```python
#Seaborn Visulization
df3 = df.head();
import seaborn as sns
sns.barplot(x='Date',y='Covaxin',data=df3)
```
